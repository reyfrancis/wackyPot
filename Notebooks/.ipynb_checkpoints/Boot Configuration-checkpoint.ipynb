## Inspecting the Nucleo Boards

As we have known, the nucleo boards has their own board stlink programmer. We will see the wirings of the boot 0 and boot 1. From the one of the nucleo boards STM32F4xx

![STM32F4_nucleo.PNG](Boot_Config_Images/STM32F4_nucleo.PNG)]

We can see that the boot0 is tied to the ground and is also tied to the morpho connector. Usually when uploading a code to the nucleo boards, we don't use any jumper connectors; hence, the boot0 must be grounded with some resistor as shown in the diagram.

Next, we can look at the boot0 of a chinese black pill. 

![boot0_blackpill.PNG](Boot_Config_Images/boot0_blackpill.PNG)

From here, we can see the same configuration. It is tied to the ground and with some jumper connectors.
